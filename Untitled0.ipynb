## ðŸŒ± ***Training And Testing Data*** ðŸŒ±
---

â €

The data we use for one process and another should not be the same, if that is the case, our predictions may be negatively affected. When we perform training and testing it is important to provide the model with new data that it is not familiar with.

To achieve the desired results, our labels should be divided into 2, in this case, our 150 flowers. The first part will compose our training-set and the second part our testing-set.

* ***ðŸ¥ž train_test_split:*** It is a scikit-learn function that separate the dataset for us, this destinates 75% as training set and 25% as testing set.

X= our data, the input of a function.

y= our labels, the output.

````
from sklearn.model_selection import train_test_split            #having sklearn, now we are going to import the train_test_split in order to use it.

X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)
````

The separation is done randomly in order to provide variety in the data. Below are the data provided to train and test respectively.

````
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
````

## ðŸ‚ ***k-Nearest Neighbors*** ðŸ‚
---

â €

A supervised classification method known as k-nearest neighbors, we will store the training-set, when a new data entered, it will be compared with the closest data and assigned its label. The k means that they can be more than one neighbor, however in this example we work only with 1.

â €

* ***Estimator classes:*** the classes of the machine learning models in scikit-learn.

â €

We need to import the k-nearest neighbors classification algorithm implemented in the KNeighborsClassifier class in the neighbors module and we need to define the number of neighbors, which is set as 1.

````
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
````

The fit method gives us a string representation of our classifier showing us the parameters used in the model.

````
knn.fit(X_train, y_train)
````

## ðŸŒº***Making Predictions & Evaluating the Model*** ðŸŒº
---

â €

"*Imagine we found an iris in the wild with a sepal length of
5 cm, a sepal width of 2.9 cm, a petal length of 1 cm, and a petal width of 0.2 cm.*"

Using a NumPy array we can calculate the shape.

````
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape: {}".format(X_new.shape))
````

Now we make the prediction using the knn object:

````
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted target name: {}".format(
iris_dataset['target_names'][prediction]))
````

The result that it brings us said that the iris specie is "setosa". In order to prove if it is true, we start the testing process with the data that was not used.

â €

We can see the predictions as well as the accuracy.

````
y_pred = knn.predict(X_test)
print("Test set predictions:\n {}".format(y_pred))

print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))

print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
````

There is also a score method of the knn object.

# ðŸµ ***Summary*** ðŸµ
---

â €

We have 3 species of flowers, each with sepals and petals of a certain length and width. To be able to classify to which species the introduced iris belongs, we consider:

â €

* ***classes:*** The species.

* ***Labels:*** The specific specie of an iris.

â €

The iris data-set contains two Numpy arrays that consist in our inputs(x) and outputs(y).

â €

The data-set needs to be split into a training set and a testing set in order to avoid the "memorization" of the data, making possible identify new data.

â €

Using k-nearest neighbors we can do the classification and prediction of our data,then we built the model by calling the fit method, and we evaluated the model using the score method.

â €

````
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
````

ðŸœ ***And our test is 97% accurate.***
